```python
import h5py
import numpy as np
import pandas as pd
```

```python
path = 'Datasets/train.hdf5/train.hdf5'
```

```python
# reads the file

file = h5py.File(path, 'r')
```

```python
# check all the groups

file.keys()
```

```python
# access the first group or in this case the only group 

group1 = file['table']
```

```python
group1.keys()
```

```python
# access deeper

group2 = group1['columns']
```

```python
group2.keys()
```

```python
customer_ID = group2['customer_ID']
customer_ID
```

```python
for name in customer_ID:
    print(name)
```

```python
# HDF5 acts like a directory or folder in computers

group2.name
```

```python
file.name
```

```python
# can iterate to show all the members

for name in file:
    print(name)
```

```python
for name in group1:
    print(name)
```

```python
# checks the metadata
file.attrs.keys()
```

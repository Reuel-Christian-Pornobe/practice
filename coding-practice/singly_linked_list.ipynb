```python
class ListNode:
    def __init__(self, val, next_node=None):
        self.val = val
        self.next = next_node

class LinkedList:
    def __init__(self):
        # Dummy node
        self.head = ListNode(-1)
        self.tail = self.head

    def get(self, index: int):
        curr = self.head.next
        i = 0
        
        while curr:
            if i == index:
                return curr.val
            i += 1
            curr = curr.next
        return -1 # Index out of bounds
    
    def insertHead(self, val):
        new_node = ListNode(val)
        new_node.next = self.head.next
        self.head.next = new_node

        if not new_node.next:         # if the next node is empty
            # If list was empy before inserting
            self.tail = new_node 
            
    def insertTail(self, val: int):
        self.tail.next = ListNode(val)
        self.tail = self.tail.next
    
    def remove(self, index):
        i = 0
        curr = self.head # start to the dummy node
        
        while i < index and curr:
            # move curr to node before target node because of the less than
            i += 1
            curr = curr.next

    
        if curr and curr.next:
            if curr.next == self.tail:
                self.tail = curr
            curr.next = curr.next.next    # remove the node and connect the nodes
            return True
        return False

    def getValues(self):
        curr = self.head.next
        res = []
        while curr:
            res.append(curr.val)
            curr = curr.next
        return res


        
```
